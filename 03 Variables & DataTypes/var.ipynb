```r
# Data types in R
# 1. Numeric
price <- 99.99
print(price)
print(class(price))

# 2. Integer
age <- 20L
print(age)
print(class(age))

# 3. Character
name <- "Laksh"
print(name)
print(class(name))

# 4. Logical
is_student <- TRUE
print(is_student)
print(class(is_student))

# Complex
z <- 1 + 2i
print(z)
print(class(z))
```

```r
# Conversion of data types
num <- 10
print(num)
print(class(num))
char_num <- as.character(num)
print(char_num)
print(class(char_num))

char <- "100"
print(char)
print(class(char))
num_char <- as.numeric(char)
print(num_char)
print(class(num_char))
```
